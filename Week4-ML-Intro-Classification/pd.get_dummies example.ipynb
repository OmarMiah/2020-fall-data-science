```python
import pandas as pd
```

```python
df = pd.DataFrame([('Foreign Cinema', 289.0),
                   ('Liho Liho', 224.0),
                   ('500 club', 80.5),
                   ('Foreign Cinema', 25.30)],
                  columns=('name','Amount')
                 )

df
```

```python
# turn it from categories to indicator columns
pd.get_dummies(df, columns=['name'])

# so at glance this function turns my data in each row of a certain column into a column 
# so my movie cinema name column above was turned into discrete columns below
# 0 and 1 indicates that i'm referencing the column or not referencing the column i'm in 0 is not 1 is yes
```

```python
#changing the name of my dummy column
pd.get_dummies(df, columns=['name'], prefix = "dmy")
```

```python
# prefix separator removes the under score 
#changing the name of my dummy column
pd.get_dummies(df, columns=['name'], prefix = "dmy", prefix_sep="*")
```

```python
# specify a parameter drop first which drops the first indicator column 
pd.get_dummies(df, columns=['name'], drop_first=True)
# drops 500 club 
# 
```
